```python
# handle activity columns(wishlist, product review etc) (fill 0 instead of NaN)
df['Wishlist_Items'] = df['Wishlist_Items'].fillna(0)
```

```python
# handle product review columns:
df['Product_Reviews_Written'] = df['Product_Reviews_Written'].fillna(0)
```

```python
# handle Mobile_App_Usage column and fill blank values wisely:

# first check how many rows are blank in Mobile_App_Usage column
print(df['Mobile_App_Usage'].isnull().sum())

# Step 1 : Use Indicator
df['missing_Mobile_App_Usage'] = df['Mobile_App_Usage'].isnull().astype(int)

# Step 2: Fill with median
df['Mobile_App_Usage'] = df['Mobile_App_Usage'].fillna(df['Mobile_App_Usage'].median())
```

```python
# handle social media engagement score column wisely:

# check missing percentage 
missing_percentage = df['Social_Media_Engagement_Score'].isnull().sum()/len(df) * 100
print(missing_percentage)

# Step 1 : Indicator
df['missing_Social_Media_Engagement_Score'] = df['Social_Media_Engagement_Score'].isnull().astype(int)

# Step 2 : fill with median
df['Social_Media_Engagement_Score'] = df['Social_Media_Engagement_Score'].fillna(df['Social_Media_Engagement_Score'].median())
```

```python
# 2nd phase: let's handle behavioral columns like session duration avg, pages_per_session, days_since_last_purchase, Email_open_rate

# first check how much percentage is blank in session_duration_avg column, pages_per_session, days_since_last_purchase, Email_open_rate......
blank_percentage_session_duration_avg = df['Session_Duration_Avg'].isnull().sum()/len(df) * 100
blank_pages = df['Pages_Per_Session'].isnull().sum()/len(df) * 100
blank_days_since = df['Days_Since_Last_Purchase'].isnull().sum()/len(df) * 100
blank_email_open = df['Email_Open_Rate'].isnull().sum()/len(df) * 100
print(blank_percentage_session_duration_avg)
print(blank_pages)
print(blank_days_since)
print(blank_email_open)
```

```python
# create indicators for missing values in behavior columns :
df['Session_missing'] = df['Session_Duration_Avg'].isnull().astype(int)
df['Pages_missing'] = df['Pages_Per_Session'].isnull().astype(int)
df['Days_missing'] = df['Days_Since_Last_Purchase'].isnull().astype(int)
df['Email_missing'] = df['Email_Open_Rate'].isnull().astype(int)
```

```python
# Step 2: Group-based Median : Use country + gender to doing group based median
df['Session_Duration_Avg'] = df.groupby(['Country','Gender'])['Session_Duration_Avg'].transform(lambda x: x.fillna(x.median()))
df['Pages_Per_Session'] = df.groupby(['Country','Gender'])['Pages_Per_Session'].transform(lambda x: x.fillna(x.median()))
df['Days_Since_Last_Purchase'] = df.groupby(['Country','Gender'])['Days_Since_Last_Purchase'].transform(lambda x: x.fillna(x.median()))
df['Email_Open_Rate'] = df.groupby(['Country','Gender'])['Email_Open_Rate'].transform(lambda x: x.fillna(x.median()))
```

```python
# Step 3: Fallback: Some groups may still have NaN values: 
df['Session_Duration_Avg'] = df['Session_Duration_Avg'].fillna(df['Session_Duration_Avg'].median())
df['Pages_Per_Session'] = df['Pages_Per_Session'].fillna(df['Pages_Per_Session'].median())
df['Days_Since_Last_Purchase'] = df['Days_Since_Last_Purchase'].fillna(df['Days_Since_Last_Purchase'].median())
df['Email_Open_Rate'] = df['Email_Open_Rate'].fillna(df['Email_Open_Rate'].median())
```

```python
# IQR CAPPING to manage Outliers in Session_duration_avg, Pages_per_session, Email_Open_rate:
cols = ['Session_Duration_Avg', 'Pages_Per_Session', 'Email_Open_Rate']
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python
# 3RD PHASE : DEAL WITH RISK AND FINANCIAL COLUMNS LIKE DISCOUNT, RETURN MISSING, CREDIT_BALANCE:

# check blank rows in Credit_Balance
blank_Credit_Balance = df['Credit_Balance'].isnull().sum()/len(df) * 100
print(blank_Credit_Balance)

# check both blank percentage of discount and return rate columns :
blank_discount = df['Discount_Usage_Rate'].isnull().sum()/len(df) * 100
blank_return_rate = df['Returns_Rate'].isnull().sum()/len(df) * 100
print(blank_discount, blank_return_rate)
```

```python
# Manage and combine logic + Statistics: 

# Discount Column:
df['Discount_Usage_Rate'] = df['Discount_Usage_Rate'].fillna(0)


# Indicator: Returns_Rate
df['missing_return_rate'] = df['Returns_Rate'].isnull().astype(int)

# Median : Returns_Rate
df['Returns_Rate'] = df['Returns_Rate'].fillna(df['Returns_Rate'].median())


# CREDIT BALANCE: INDICATOR + FILL 0.
df['missing_credit'] = df['Credit_Balance'].isnull().astype(int)
df['Credit_Balance'] = df['Credit_Balance'].fillna(0)
```

```python
# Apply IQR CAPPING ON THESE FINANCIAL COLUMNS:
cols = ['Discount_Usage_Rate','Returns_Rate', 'Credit_Balance']
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 - 1.5 * IQR

    df[col] = df[col].clip( lower, upper )
```

```python
# Handle Customer_Service_Calls & Payment_Method_Diversity:


# Customer Service call:
df['Customer_Service_Calls'] = df['Customer_Service_Calls'].fillna(df['Customer_Service_Calls'].median())


#Payment_Method_Diversity:
# make indicator
df['Payment_Method_Diversity'] = df['Payment_Method_Diversity'].isnull().astype(int)

# fill median with groupby(country)
df['Payment_Method_Diversity'] = df.groupby('Country')['Payment_Method_Diversity'].transform(lambda x: x.fillna(x.median()))
df['Payment_Method_Diversity'] = df['Payment_Method_Diversity'].fillna(df['Payment_Method_Diversity'].median())
```
